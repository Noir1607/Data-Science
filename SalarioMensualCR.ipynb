```python
def calcular_total_empleo(df, año):
    trimestres = ["EFM", "AMJ", "JAS", "OND"]
    columnas = [f"{trimestre} {año}" for trimestre in trimestres]

    if not all(col in df.columns for col in columnas):
        raise ValueError(f"No se encontraron todas las columnas necesarias para el año {año}")

    # Sumamos los valores de los trimestres y dividimos entre 4 para obtener el promedio anual
    df[f"Total{año}"] = df[columnas].sum(axis=1) / 4
    return df
```

```python
años = [ 2011, 2012, 2013, 2014, 2015, 2016 , 2017 ,2018 ,  2019 , 2020 ,2021 , 2022 , 2023]
```

```python
for año in años:
    df = calcular_total_empleo(df, año)
```
