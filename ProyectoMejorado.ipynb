```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/netflix_titles.csv")
```

```python
df
```

```python
df.info()
```

```python
df = df.drop("show_id", axis = 'columns')
```

```python
columnas = {
            "type" : "Type",
            "title" : "Title",
            "director":"Director",
            'cast' : "Cast",
            'country': "Country",
            'date_added' : "Date_Added",
            "release_year" : "Release_Year",
            "rating" : "Rating",
            "duration" : "Duration",
            "listed_in" : "Genre",
            "description":"Description"
}

df.rename(columns=columnas, inplace=True)
```

```python
df
```

```python
df["Type"].value_counts()
```

```python
import datetime
```

```python
df['Date_Added'] = pd.to_datetime(df['Date_Added'], format="%B %d, %Y", errors='coerce')
df
```

```python

normalized_genres = df['Genre'].str.split(',')

normalized_genres = normalized_genres.apply(lambda x: [genre.strip() for genre in x])

genre_counts = pd.Series([genre for sublist in normalized_genres for genre in sublist]).value_counts()
```

```python
genre_counts_df = genre_counts.reset_index()
genre_counts_df.columns = ['Genre', 'Count']
genre_counts_df.head()
```

```python
filtro_generos = df[df['Genre'].str.contains('International Movies', case=False)]
filtro_generos
```

```python
df['Type'].value_counts().plot(kind='pie', colors=['lightblue', 'lightcoral']).set_title("Grafico de comparacion")
```

```python
import matplotlib.pyplot as plt

primeras_cinco = genre_counts_df.head(5)

plt.figure(figsize=(8, 8))
plt.pie(primeras_cinco['Count'], labels=primeras_cinco['Genre'], colors=['skyblue', 'lightcoral', 'lightgreen', 'orange', 'pink'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribución de los 5 primeros géneros')
plt.show()
```
