```python
hist1 = df['Costo Por Unidad']
```

```python
plt.figure(figsize=(8, 6))
plt.hist(hist1, bins=20, edgecolor='black')
plt.title(f'Histograma de Costo por Unidad')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
hist1.value_counts()
```

Aqui se refleja , todos loa valores de el costo por unidad y en las frecuencias se refleja la frecuencia de la repeticion de los datos , vease que por ejemplo, el monto de 250.000 se repite 1667 veces , cosa que se refleja en el histograma

```python
columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns

# Crear un gráfico para cada variable numérica
for columna in columnas_numericas:
    plt.figure(figsize=(8, 6))
    df[columna].hist(bins=30, edgecolor='black')
    plt.title(f'Histograma de {columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.show()
```
