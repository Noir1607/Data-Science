#Importar Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Importar informacion

```python
df = pd.read_csv('/content/FG_500AC.csv', sep=';', encoding='latin-1', na_values="-")
df.head()
```

#Limpieza de datos && Reemplazo de " , " , "$" y "%"

```python
df['Number of Employees'] = df['Number of Employees'].str.replace(' ','')
df.head()
```

```python
df['Revenues(millions)'] = df['Revenues(millions)'].str.replace(' ','')
df.head()
```

```python
 df['Revenue Change'] = df['Revenue Change'].str.replace('%','')
 df.head()
```

```python
df['Profits(millions)'] = df['Profits(millions)'].str.replace(' ','')
df.head()
```

```python
df['Profit Change'] = df['Profit Change'].str.replace('%','')
df.head()
```

```python
df['Assets (millions)'] = df['Assets (millions)'].str.replace(' ','')
df.head()
```

#Vamos a cambiar los dtypes

```python
df.dtypes
```

```python
df['Number of Employees'] = df['Number of Employees'].astype('int64')
df.info()
```

```python
df['Revenues(millions)'] = df['Revenues(millions)'].replace('kkkk', np.nan)
df['Revenues(millions)'] = df['Revenues(millions)'].astype('float64')
df['Revenues(millions)'] = df['Revenues(millions)'].fillna(0)
df.info()
```

```python
df['Revenue Change'] = df['Revenue Change'].str.replace(',','.')
df['Revenue Change'] = df['Revenue Change'].astype('float64')
df.info()
```

```python
df['Profits(millions)'] = df['Profits(millions)'].replace('ptt', np.nan).replace('xtl', np.nan)
df['Profits(millions)'] = df['Profits(millions)'].astype('float64')
df['Profits(millions)'] = df['Profits(millions)'].fillna(0)
df.info()
```

```python
df['Profit Change'] = df['Profit Change'].str.replace(',','.')
df['Profit Change'] = df['Profit Change'].astype('float64')
df.info()
```

```python
df['Assets (millions)'] = df['Assets (millions)'].astype('float64')
df.info()
```

```python
df.head()
```

##Un poquito mas de limpieza de paises para que no salgan repetidos

```python
variantes_eu = ['Estados Unidos','U.S.','US','EstadosUnidos','U.S.A.']

reemplazos = {variante: 'United States' for variante in variantes_eu }

df['Country'] = df['Country'].replace(reemplazos)

paises_unicos = df['Country'].unique()

for i in paises_unicos:
    print(i)
```

```python
variantes_ch = ['CHINA' , 'Chinnise']

reemplazos = {variante: 'China' for variante in variantes_ch }

df['Country'] = df['Country'].replace(reemplazos)

paises_unicos = df['Country'].unique()

for i in paises_unicos:
    print(i)
```

##Graficos

```python

empleados_por_pais = df.groupby('Country')['Number of Employees'].sum().reset_index()

empleados_por_pais = empleados_por_pais.sort_values(by='Number of Employees', ascending=True)


# Crear el grfico de barras
plt.figure(figsize=(10, 6))
plt.barh(empleados_por_pais['Country'], empleados_por_pais['Number of Employees'],edgecolor = 'black' , color = 'slateblue')
plt.xlabel('Cantidad de Empleados')
plt.ylabel('Paises')
plt.title('Cantidad de empleados por pas')

# Mostrar el grfico
plt.show()
```

1. China y Estados Unidos encabezan la lista con la mayor cantidad de empleados, muy por encima del resto de pases.
2. Pases como Japn, Alemania, Francia y Reino Unido tambin tienen cantidades considerables de empleados, aunque significativamente menores que China y Estados Unidos.
3. Se observan varios pases europeos como Pases Bajos, Rusia, Suiza y Canad con niveles de empleo relativamente altos.
4. Hay una serie de pases, incluyendo algunos asiticos como Corea del Sur, India y Singapur, as como europeos como Espaa, Irlanda e Italia, con niveles de empleo intermedios.
5. En la parte inferior de la grfica, se encuentran pases con menores cantidades de empleados, como Finlandia, Noruega, Austria y Tailandia, entre otros.

```python
millones_por_pais = df.groupby('Country')['Revenues(millions)'].sum().reset_index()

millones_por_pais = millones_por_pais.sort_values(by='Revenues(millions)', ascending=False)

# Crear el grfico de barras
plt.figure(figsize=(10, 6))
plt.barh(millones_por_pais['Country'], millones_por_pais['Revenues(millions)'],edgecolor = 'black' , color = 'orange')
plt.xlabel('Ganancia en Milllones')
plt.ylabel('Paises')
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x
plt.title('Cantidad de millones ganados por pas')

# Mostrar el grfico
plt.show()
```

1. Estados Unidos y China encabezan la lista con las mayores ganancias en millones, muy por encima del resto de los pases.
2. Pases como Japn, Alemania, Reino Unido y Corea del Sur tambin presentan ganancias considerables, aunque significativamente menores que Estados Unidos y China.
3. Se observa un grupo de pases europeos como Suiza, Pases Bajos, India, Canad y Taiwn con niveles de ganancias intermedios.
4. Varios pases rabes, como Arabia Saudita, Rusia e Italia, tambin se ubican en el rango medio de ganancias.
5. En la parte inferior de la grfica se encuentran pases con menores ganancias, como Finlandia, Polonia, Turqua y Austria, entre otros.
